```python
import sys
import os
import pymongo 
from pymongo import MongoClient
import pprint
mongodb_host='localhost'
mongodb_port='27017'
client = MongoClient ('mongodb://localhost:27017/')
db=client['db']
c1=db.pp2_0
```

#### We got data from Twitter API.  In Python, we created a list for 3 tags, which are 3 football players' name.  And we used Twitter API  with  Python code to get the related data.

<font color=3E2F2C size=4 face="Didot"><B>1. What are people saying about me (somebody)?  

#### Because the data we scrapied is about three tags.  For this question,  we used  Mongodb to get all the data in dataset. They are alll related the tags. So they can completely explain what are people saying about me.

```python
for post in db.pp2_0.find({}):
    pprint.pprint(post)
```

<font color=#3E2F2C size=4 face="Didot"><B>2. How viral are my posts? </font>

#### We suppose that the retweet number an reflect how viral of every post. So in this query,  we ordered the retweet count in Mongodb to show all posts virality. And then just show users screen_name and its related retweet number.   

```python
for post in db.pp2_0.find({},{'_id':0,'screen_name':1,'retweet_count':1,}).sort('retweet_count',-1):
    pprint.pprint(post)
```

<font color=#3E2F2C size=4 face="Didot"><B>3. What topics are trending in my domain? </font>

#### Because the list we created to scrapied the data is in three name. In this question, we filtered  the name of  each one in the list and constrained in one same minute like 14:28 to count which one has the most tweets. From the outcome below, we can conclude Cristiano Ronaldo is most trending.  

```python
db.pp2_0.find({'$and':[{'create_time':{'$regex': '14:28'}},{'text':{'$regex':'Lionel Messi'}}]}).count()
```

```python
db.pp2_0.find({'$and':[{'create_time':{'$regex':'14:28'}},{'text':{'$regex':'Cristiano Ronaldo'}}]}).count()
```

```python
db.pp2_0.find({'$and':[{'create_time':{'$regex':'14:28'}},{'text':{'$regex':'Neymar'}}]}).count()
```

<font color=#3E2F2C size=4 face="Didot"><B>4. What posts are like mine? </font>

#### In this question, we assume people with same tags are like others. So, we randomly selected one like 'Cristiano Ronaldo'.  And we get the user  screen_name and text of users who posted tags in 'Cristiano Ronaldo'. 

```python
for post in db.pp2_0.find({'text':{'$regex':'Cristiano Ronaldo'}},{'screen_name':1,'_id':0,'text':1}):
  pprint.pprint(post)
```

<font color=#3E2F2C size=4 face="Didot"><B>5.What users post like me? </font>

#### In this question, we assume people with same tags are like others. For example, people posted with 'Cristiano Ronaldo' tags are like others. Then we get their id and screen_name.

```python
for post in db.pp2_0.find({'text':{'$regex':'Cristiano Ronaldo'}},{'id':1,'screen_name':1, '_id':0}):
  pprint.pprint(post)
```

<font color=#3E2F2C size=4 face="Didot"><B>6. Who should I be following? </font>

####  We assume user we searched in API were all taged in 'Cristiano Ronaldo'.   And we think they are likely to follow people  with the a great amont of  followers count. So we put them in a descending order to see which one they are most interested in following. 

```python
for post in db.pp2_0.find({'text':{'$regex':'Cristiano Ronaldo'}},{'screen_name':1,'followers_count':1, '_id':0}).sort('followers_count',-1):
  pprint.pprint(post)
```

<font color=#3E2F2C size=4 face="Didot"><B>7. What keywords/ hashtags should I add to my post? </font>

#### We count all the data in a same period of different names. We seperated it in three times, because we have three names. From the query below, we can get the 'CR' ,'LM', 'NM' names appeares 48, 10,14 times, respectively. 

```python
db.pp2_0.find({'text':{'$regex':'Cristiano Ronaldo'}}).count()
```

```python
db.pp2_0.find({'text':{'$regex':'Lionel Messi'}}).count()
```

```python
db.pp2_0.find({'text':{'$regex':'Neymar'}}).count()
```

<font color=#3E2F2C size=4 face="Didot"><B>8.How much influence to my posts have? </font>

#### We assume that number of retweets would show how much influence one post have. So we used id, screen_name and retweet_counts to explain posts' influence and their related users.

```python
for post in db.pp2_0.find({'text':{'$regex':'Cristiano Ronaldo'}},{'id':1,'screen_name':1,'_id':0,'retweet_count':1}):
  pprint.pprint(post)
```

<font color=#47413E size=5 face="Didot"><B>SQL VS NoSQL </font>

<font color=#47413E size=4 face="Didot"><B>SQL databases are table based databases whereas NoSQL databases are document based, key-value pairs, graph databases or wide-column stores. This means that SQL databases represent data in form of tables which consists of n number of rows of data whereas NoSQL databases are the collection of key-value pair, documents, graph databases or wide-column stores which do not have standard schema definitions which it needs to adhered to. </font>
